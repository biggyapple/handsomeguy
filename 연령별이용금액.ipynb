```python
import pandas as pd
agedata=pd.read_excel('C:/Users/user/Downloads/연령별데이터.xlsx')
del agedata["시도명"]
del agedata["데이터기준일자"]
del agedata["업종코드"]
del agedata["매장수"]
del agedata["지역구분"]
del agedata["성별"]
del agedata["업종명 대분류"]
del agedata["관광구분"]
del agedata["이용자 구분"]
pd.options.display.float_format = '{:.2f}'.format
agedata=agedata.dropna()

agedata
```

```python
#43 개 동분리
select = agedata["읍면동명"].isin(['건입동', '구좌읍', '남원읍', '노형동','대륜동', '대정읍', '대천동', '도두동', '동홍동', '봉개동',
       '삼도1동', '삼도2동', '삼양동', '서홍동', '성산읍', '송산동', '아라동', '안덕면', '애월읍',
       '연동', '영천동', '예래동', '오라동', '외도동', '용담1동', '용담2동', '우도면', '이도1동',
       '이도2동', '이호동', '일도1동', '일도2동', '정방동', '조천읍', '중문동', '중앙동', '천지동',
       '추자면','표선면', '한경면','한림읍', '화북동', '효돈동'])
agedata = agedata[select]
agedata
```

```python
#소매업으로 분류
some=agedata[(agedata['업종명']=="과실 및 채소 소매업")
    |(agedata["업종명"]=="관광 민예품 및 선물용품 소매업")
   |(agedata["업종명"]=="수산물 소매업")
    |(agedata["업종명"]=="육류 소매업")
    |(agedata["업종명"]=="빵 및 과자류 소매업")
    |(agedata["업종명"]=="화장품 및 방향제 소매업")]
#년도별 합치기
some_group=some.groupby(["읍면동명","연령대","이용금액"],as_index=False).sum()
del some_group["업종명"]
some_group["업종명"]="소매업"

some_group

       
```

```python
some_group_1020=some_group[(some_group['연령대']==10.0)
                           |(some_group["연령대"]==20.0)]
some_group_1020["연령대"] = 1020

some_group_3050=some_group[(some_group["연령대"]=="30")
                          |(some_group["연령대"]=="40")
                          |(some_group["연령대"]=="50")]

some_group_6070=some_group[(some_group["연령대"]=="60.0")
                          |(some_group["연령대"]=="70.0")]

some1020=some_group_1020.groupby(["읍면동명","업종명"],as_index=False).mean()                        
some1020.columns=['읍면동명','업종명','연령대',"1인당 이용금액"]
some1020
```
